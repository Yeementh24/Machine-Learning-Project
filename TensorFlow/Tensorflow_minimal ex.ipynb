```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observation=1000
xs=np.random.uniform(-10,10,size=(observation,1))
zs=np.random.uniform(-10,10,size=(observation,1))

generated_inputs=np.column_stack((xs,zs))

noise=np.random.uniform(-1,1,(observation,1))

generated_target=2*xs-3*zs+5+noise

np.savez('TF_intro',inputs=generated_inputs,targets=generated_target)
```

```python
training_data= np.load('TF_intro.npz')
```

```python
input_size=2
output_size=1

model= tf.keras.Sequential([ 
    tf.keras.layers.Dense(output_size,
                         kernel_initializer=tf.random_uniform_initializer(-0.1,0.1),
                         bias_initializer=tf.random_uniform_initializer(-0.1,0.1))
    
                           ])

custom_opyimizer=tf.keras.optimizers.SGD(learning_rate=0.02)

model.compile(optimizer=custom_opyimizer,loss='mean_squared_error')
model.fit(training_data['inputs'],training_data['targets'],epochs=100,verbose=2)
```

```python
model.layers[0].get_weights()
```

```python
weights=model.layers[0].get_weights()[0]
print(weights)
bias=model.layers[0].get_weights()[1]
print(bias)
```

```python
model.predict_on_batch(training_data['inputs']).round(1)
```
