### Multicollinearity

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables=data_cleaned[['Mileage','Year','EngineV']]
vif=pd.DataFrame()
vif['VIF']=[variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif['features']=variables.columns
```

```python
vif
```

- VIF=1 : no multicollinearity
- 1 < VIF < 5 : perfectly okay
- 5 < VIF : unacceptable

- therefor ['Year'] show to much collinearity in the data so we will drop the ['Year'] Column

```python
data_no_multicolli=data_cleaned.drop(['Year'],axis=1)
```

 ### create dummy variables

```python
data_with_dummies=pd.get_dummies(data_no_multicolli,drop_first=True)
```

```python
cols=['log_price','Mileage',
 'EngineV',
 'Brand_BMW',
 'Brand_Mercedes-Benz',
 'Brand_Mitsubishi',
 'Brand_Renault',
 'Brand_Toyota',
 'Brand_Volkswagen',
 'Body_hatch',
 'Body_other',
 'Body_sedan',
 'Body_vagon',
 'Body_van',
 'Engine Type_Gas',
 'Engine Type_Other',
 'Engine Type_Petrol',
 'Registration_yes']
```

```python
data_preprocessed=data_with_dummies[cols]
data_preprocessed.head()
```

## linear regression model

### Declare the input and the Targets

```python
targets=data_preprocessed['log_price']
inputs=data_preprocessed.drop(['log_price'],axis=1)
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler


scaler=StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled=scaler.transform(inputs)
```

- it is not usually recommended to standardize dummy variables

- Scaling has no effect on the predictive power of dummies once scaled,though they lose all their dummy meaning 


### Train Test Split

```python
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(inputs_scaled,targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_hat=reg.predict(X_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('target(y_train)',size=18)
plt.ylabel('prediction(y_hat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.displot(y_train-y_hat,kde=True)
plt.title('Residuals PDF',size=18)
```

- there are certain observation for which (y_train-y_hat) is much lower than 
  the mean (a much higher price is predicted than is observed)

```python
reg.score(X_train,y_train)
```

### Finding the weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary=pd.DataFrame(inputs.columns.values,columns=['Features'])
reg_summary['Weights']=reg.coef_
reg_summary
```

```python
data_cleaned['Brand'].unique()
```

### Weight Interpretation
####   I. Continuous Variables:
-    1. A positive weight shows that as a feature increase in value, so do the log_price and price respectively
-    2. a negative weight shows that as a feature increase in value, so do the log_price and price decrease

####   II. Dummy Variables:
-    1. A positive weight shows that the respective catogry ( brand ) is more expensive than the benchmark(Audi)
-    2. A positive weight shows that the respective catogry ( brand ) is less expensive than the benchmark(Audi)

##### THe Bigger the weight the Bigger the impact

## Testing the model

```python
y_hat_test =reg.predict(X_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel('target(y_test)',size=18)
plt.ylabel('prediction(y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

- we have a higher concentration of values around the 45 degree line
- we can say that our model is very good at predicting higher prices
- but for the lower one stutation is not so good as they are more scatter pointing to the fact that we are not geting the price right


```python
df_pf=pd.DataFrame(np.exp(y_hat_test),columns=['Prediction'])
df_pf.head()
```
